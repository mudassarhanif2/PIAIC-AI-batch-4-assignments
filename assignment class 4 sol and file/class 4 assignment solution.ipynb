```python
# declare an empty list

lst = []

# declare a dictionary with keys only

dic = {
    "name" : "",
    "age" : "",
    "work" : "",
    "skills" : "",
    "degree" : "",
    "salary" : "",
}

usr = int(input("How many users do you want to Enter : "))

for i in range(usr):
    print(f"Enter Data of user {i+1}\n\n")
    name = input("Enter user name : ")
    dic["name"] = name
    age = int(input("Enter user age : "))
    dic["age"] = age
    work = input("Enter user work : ")
    dic["work"] = work
    print("Enter any of user 3 skills : ")
    
    #declare an empty list for skills
    skill = []
    
    for j in range(3):
        usr_skill = input(f"{j+1}---> ")
        skill.append(usr_skill)
    
    dic["skills"] = skill
    
    #declare a dictionary for degree details
    
    deg = {
        "title" : "",
        "major" : "",
        "completion date" : "",
    }
    
    title = input("Enter title of your degree : ")
    deg["title"] = title
    major = input("Enter your major subject : ")
    deg["major"] = major
    cmpl_date = input("Enter completion date of your degree : ")
    deg["completion date"] = cmpl_date
    
    dic["degree"] = deg
    
    salary = int(input("Enter your salary : "))
    dic["salary"] = salary
    
    lst.append(dic)
    
print(lst)
```
